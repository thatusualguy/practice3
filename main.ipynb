```python

# 2. Тепловая карта корреляций между разными аспектами курсов
fig, axes = plt.subplots(2, 2, figsize=(18, 15))
fig.suptitle('Корреляции между оценками курсов', fontsize=16, fontweight='bold')

course_names = ['АД','ИСУ','БД','ВССИТ','ОТИС','СОИ','РСПСИТ']
courses = ['ad', 'isu', 'bd', 'vssit', 'otis', 'soi', 'rspsit']

# Функция для создания тепловой карты определенных метрик
def plot_correlation_heatmap(metric, ax, title):
    metric_cols = [col for col in df.columns if metric in col and df[col].dtype in ['int64', 'float64']]
    if metric_cols:
        corr_matrix = df[metric_cols].corr()
        # Маска для верхнего треугольника
        mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

        sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f',
                   cmap='coolwarm', center=0, ax=ax,
                   square=True, linewidths=1, cbar_kws={"shrink": 0.8})
        ax.set_title(title, fontsize=12, fontweight='bold')

        # Поворот подписей
        ax.set_xticklabels([course_names[(courses.index(col.replace(f'{metric}_', '')))] for col in metric_cols],
                          rotation=45, ha='right')
        ax.set_yticklabels([course_names[(courses.index(col.replace(f'{metric}_', '')))] for col in metric_cols],
                          rotation=0)

# Интерес к курсам
plot_correlation_heatmap('interest', axes[0,0], 'Корреляция интереса к разным курсам')

# Сложность курсов
plot_correlation_heatmap('difficulty', axes[0,1], 'Корреляция сложности разных курсов')

# Современность курсов
plot_correlation_heatmap('modern', axes[1,0], 'Корреляция оценки современности курсов')

# Оценка преподавателей
plot_correlation_heatmap('teacher', axes[1,1], 'Корреляция оценок преподавателей')

plt.tight_layout()
plt.show()
```

## Фридман

```python
from scipy.stats import friedmanchisquare

def fridnam(ranks):
    stat, p = friedmanchisquare(*ranks.values)
    return stat, p
```

```python
def kendall_concordance(df):
    """
    Вычисляет коэффициент конкордации Кендалла W
    для всех экспертов одновременно

    df: DataFrame, где строки - эксперты, столбцы - объекты
    """
    n = df.shape[0]  # количество экспертов
    m = df.shape[1]  # количество объектов

    # 1. Вычисляем сумму рангов для каждого объекта
    R = df.sum(axis=0)  # сумма по столбцам (для каждого объекта)

    # 2. Вычисляем среднюю сумму рангов
    R_mean = R.mean()

    # 3. Вычисляем сумму квадратов отклонений
    S = ((R - R_mean) ** 2).sum()

    # 4. Вычисляем коэффициент конкордации
    # Формула для случая без связанных рангов
    W = 12 * S / (n**2 * (m**3 - m))

    # Ограничиваем значение от 0 до 1
    W = max(0, min(1, W))

    return float(W)
```

```python
def stats(df):
    mean_series = df.mean()
    std_series = df.std()
    var_series = df.var()

    # 2. Создаем коэффициент вариации (CV)
    cv_series = (std_series / mean_series)

    # 3. Создаем датафрейм со всеми метриками
    metrics_df = pd.DataFrame({
        'mean': mean_series,
        'std': std_series,
        'var': var_series,
        'cv': cv_series
    })
    return metrics_df
```

```python
def metrics(ranks):
    print("Фридман")
    stat, p = fridnam(ranks)
    print(f'{stat=:.2f}, {p=:.2f}')

    print('Кендалл')
    print(kendall_concordance(ranks))

    return stats(ranks)
```

```python
def display_dev():
    mean_ranks = ranks.mean()
    deviations = ranks.sub(mean_ranks, axis=1)
    avg_abs_deviation = deviations.abs().mean(axis=1)

    sorted_deviations = avg_abs_deviation.sort_values(ascending=False)

    sns.barplot(sorted_deviations.values)
    plt.xticks(range(len(sorted_deviations)), sorted_deviations.index)
    plt.title('Среднее абсолютное отклонение оценок экспертов от общего среднего')
    plt.xlabel('Эксперты')
    plt.ylabel('Среднее абсолютное отклонение, баллы')
    plt.show()

display_dev()
```

```python
import numpy as np

# Быстрый метод определения выбивающихся
def find_outliers_simple(ranks):
    # 1. Отклонение от среднего ранга
    mean_ranks = ranks.mean(axis=0)
    deviations = np.abs(ranks - mean_ranks).mean(axis=1)

    # 2. Критерий: отклонение > 75% перцентиль
    threshold = np.percentile(deviations, 75)
    outliers = np.where(deviations > threshold)[0]

    # 3. Дополнительно: респонденты с противоположными рангами
    # (ставят на первое место то, что другие на последнее)
    correlation_with_mean = []
    for i in range(len(ranks)):
        corr = np.corrcoef(ranks.iloc[i], mean_ranks)[0, 1]
        correlation_with_mean.append(corr)

    # Те, у кого отрицательная корреляция с общим мнением
    negative_corr = np.where(np.array(correlation_with_mean) < 0)[0]

    # Объединяем
    all_outliers = set(list(outliers) + list(negative_corr))

    return list(all_outliers), deviations, correlation_with_mean

# Использование
outliers, deviations, correlations = find_outliers_simple(ranks)
print(f"Выбивающихся респондентов: {len(outliers)}")
print(list(map(int,outliers)))
ranks = ranks.drop(index=outliers)
print(f"Удалено {len(outliers)} респондентов")
```

```python
def seaborn_grouped_barplot(df1, df2, value_column='std'):
    """
    Группированный барплот с использованием Seaborn
    """

    df1_copy = df1.copy()
    df2_copy = df2.copy()

    # Добавляем метку источника данных
    df1_copy['Источник'] = 'До очистки'
    df1_copy['Предмет'] = df1_copy.index

    df2_copy['Источник'] = 'После очистки'
    df2_copy['Предмет'] = df2_copy.index

    combined_df = pd.concat([df1_copy, df2_copy])

    ax= sns.barplot(
        data=combined_df,
        x='Предмет',
        y=value_column,
        hue='Источник',
        legend=True
    )

    ax.bar_label(ax.containers[0], fontsize=10, fmt='%.2f')
    ax.bar_label(ax.containers[1], fontsize=10, fmt='%.2f')

    # # Настраиваем графики
    # for ax in g.axes.flat:
    #     ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
    #     ax.grid(True, alpha=0.3, axis='y')
    #
    #     # Добавляем значения на столбцы
    #     for container in ax.containers:
    #         ax.bar_label(container, fmt='%.1f', padding=3, fontsize=9)

    # g.fig.suptitle('Сравнение данных по городам', y=1.05, fontsize=14, fontweight='bold')
    plt.title('Изменение дипсперсии после очистки')
    plt.ylabel('Дисперсия')
    plt.xticks(df1_copy['Предмет'].values, ['БД','АД','ВССИТ','РСПСИТ','СОИ','ОТИС','ИСУ'])
    plt.tight_layout()
    plt.show()

seaborn_grouped_barplot(before_experts, after_experts)
```

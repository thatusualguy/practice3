```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# Настройка стиля
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# 1. Анализ распределения ответов по группам (group)
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Распределение ответов по группам', fontsize=16, fontweight='bold')

# Распределение групп
axes[0,0].hist(df['group'], bins=20, edgecolor='black', alpha=0.7)
axes[0,0].set_xlabel('Группа')
axes[0,0].set_ylabel('Количество студентов')
axes[0,0].set_title('Распределение студентов по группам')
axes[0,0].grid(True, alpha=0.3)

# Посещаемость по группам
attendance_by_group = df.groupby('group')['attendance'].value_counts().unstack()
attendance_by_group.plot(kind='bar', stacked=True, ax=axes[0,1], alpha=0.8)
axes[0,1].set_xlabel('Группа')
axes[0,1].set_ylabel('Количество студентов')
axes[0,1].set_title('Посещаемость по группам')
axes[0,1].legend(title='Посещаемость')
axes[0,1].tick_params(axis='x', rotation=45)

# Дополнительные материалы по группам
materials_by_group = df.groupby('group')['additional_materials'].value_counts().unstack()
materials_by_group.plot(kind='bar', stacked=True, ax=axes[1,0], alpha=0.8)
axes[1,0].set_xlabel('Группа')
axes[1,0].set_ylabel('Количество студентов')
axes[1,0].set_title('Использование доп. материалов по группам')
axes[1,0].legend(title='Материалы')
axes[1,0].tick_params(axis='x', rotation=45)

# Намерение посещать разные курсы (среднее по группам)
willgo_cols = [col for col in df.columns if 'willgo_' in col]
willgo_means = df.groupby('group')[willgo_cols].mean()
willgo_means.T.plot(kind='bar', ax=axes[1,1], alpha=0.8, width=0.8)
axes[1,1].set_xlabel('Курсы')
axes[1,1].set_ylabel('Средний балл намерения')
axes[1,1].set_title('Намерение посещать курсы по группам')
axes[1,1].legend(title='Группа')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python

# 2. Тепловая карта корреляций между разными аспектами курсов
fig, axes = plt.subplots(2, 2, figsize=(18, 15))
fig.suptitle('Корреляции между оценками курсов', fontsize=16, fontweight='bold')

# Функция для создания тепловой карты определенных метрик
def plot_correlation_heatmap(metric, ax, title):
    metric_cols = [col for col in df.columns if metric in col and df[col].dtype in ['int64', 'float64']]
    if metric_cols:
        corr_matrix = df[metric_cols].corr()
        # Маска для верхнего треугольника
        mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

        sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f',
                   cmap='coolwarm', center=0, ax=ax,
                   square=True, linewidths=1, cbar_kws={"shrink": 0.8})
        ax.set_title(title, fontsize=12, fontweight='bold')

        # Поворот подписей
        ax.set_xticklabels([col.replace(f'{metric}_', '') for col in metric_cols],
                          rotation=45, ha='right')
        ax.set_yticklabels([col.replace(f'{metric}_', '') for col in metric_cols],
                          rotation=0)

# Интерес к курсам
plot_correlation_heatmap('interest', axes[0,0], 'Корреляция интереса к разным курсам')

# Сложность курсов
plot_correlation_heatmap('difficulty', axes[0,1], 'Корреляция сложности разных курсов')

# Современность курсов
plot_correlation_heatmap('modern', axes[1,0], 'Корреляция оценки современности курсов')

# Оценка преподавателей
plot_correlation_heatmap('teacher', axes[1,1], 'Корреляция оценок преподавателей')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import friedmanchisquare

# Самый простой способ
def quick_friedman(df, postfix):
    cols = [c for c in df.select_dtypes(include=np.number).columns if c.endswith(postfix)]
    if len(cols) >= 3:
        stat, p = friedmanchisquare(*[df[c].values for c in cols])
        print(f"{postfix:10} | χ² = {stat:6.2f} | p = {p:8.6f} | ", 
              end='')
        print("Различия есть" if p < 0.05 else "Различий нет")
        return stat

friedmans = {}

# Анализ всех категорий
print("КАТЕГОРИЯ      | χ²       | p-значение  | Вывод")
print("-" * 50)
for postfix in ['bd', 'ad', 'vssit', 'rspsit', 'soi', 'otis', 'isu']:
	friedmans[postfix] = quick_friedman(df, postfix)

# ['willgo', 'interest', 'profit', 'teacher', 'understand', 'modern', 'difficulty']
```

# Коэффициент конкордации Кендалла

## До удаления "плохих" респондентов

```python
import numpy as np
import pandas as pd
from scipy.stats import rankdata, chi2

def kendall_w(data, friedman):
    """Расчет коэффициента конкордации Кендалла"""
    data = np.array(data)
    m, n = data.shape  # m экспертов, n объектов
    
    # Ранжируем по строкам
    ranks = np.apply_along_axis(rankdata, 1, data)
    
    # Сумма рангов по объектам
    R = ranks.sum(axis=0)
    R_mean = R.mean()
    
    # Коэффициент W
    W = friedman / (m * (n-1))

	# Проверка значимости
    p_val = 1 - chi2.cdf(friedman, n - 1)
    
    return W, p_val

data = df.select_dtypes(include=[np.number])
kendalls = {}
for postfix in friedmans:
	cols = [c for c in data.columns if c.endswith(postfix)]

	W, p = kendall_w(data[cols].values, friedmans[postfix])

	kendalls[postfix] = W

	print(f"= {postfix} =")
	print(f"Коэффициент конкордации W = {W:.3f}")
	print(f"p-значение = {p:.4f}")

	if p < 0.05:
		print("✓ Согласованность значима (p < 0.05)")
	else:
		print("✗ Согласованность не значима")

	# Интерпретация
	if W < 0.1: print("Очень слабая")
	elif W < 0.3: print("Слабая")
	elif W < 0.5: print("Умеренная")
	elif W < 0.7: print("Сильная")
	else: print("Очень сильная")
	print("-"*20)

sns.barplot(kendalls)
```

## Удаляем выбивающихся респондентов

```python
import numpy as np

# Быстрый метод определения выбивающихся
def find_outliers_simple(df, columns):
    data = df[columns].values
    ranks = np.apply_along_axis(rankdata, 1, data)
    
    # 1. Отклонение от среднего ранга
    mean_ranks = ranks.mean(axis=0)
    deviations = np.abs(ranks - mean_ranks).mean(axis=1)
    
    # 2. Критерий: отклонение > 75% перцентиль
    threshold = np.percentile(deviations, 75)
    outliers = np.where(deviations > threshold)[0]
    
    # 3. Дополнительно: респонденты с противоположными рангами
    # (ставят на первое место то, что другие на последнее)
    correlation_with_mean = []
    for i in range(len(data)):
        corr = np.corrcoef(ranks[i], mean_ranks)[0, 1]
        correlation_with_mean.append(corr)
    
    # Те, у кого отрицательная корреляция с общим мнением
    negative_corr = np.where(np.array(correlation_with_mean) < 0)[0]
    
    # Объединяем
    all_outliers = set(list(outliers) + list(negative_corr))
    
    return list(all_outliers), deviations, correlation_with_mean

# Использование
outliers, deviations, correlations = find_outliers_simple(df, df.select_dtypes(include=[np.number]).columns)
print(f"Выбивающихся респондентов: {len(outliers)}")

clean_df = df.drop(index=outliers).reset_index(drop=True)
print(f"Удалено {len(outliers)} респондентов")
```
